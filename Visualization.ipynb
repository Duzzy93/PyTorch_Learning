```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import data
```

```python
x= np.linspace(0,10,100)
y1 = x**2
y2 = 10*x+2

x2 = 10+10*np.random.randn(1000)
```

```python
plt.plot(x, y1, 'r--', label='y1')
plt.plot(x, y2, 'b--', label='y2')
plt.title("line plot")
plt.xlabel("x")
plt.ylabel("y")
plt.legend(loc='upper left')
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x, y1, 'g--')
plt.subplot(1,2,2)
plt.plot(x, y2, '-')
plt.show()
```

```python
plt.hist(x2, bins=50, density=True, facecolor='orange')
plt.show()
```
